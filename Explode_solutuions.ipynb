```
#calculate fi(t)

def fi_t(point):
  return math.exp( -1/(point*point) )
```

```
def get_y_array_derive_k_dimension(array,k):
  if (k==0):
    return array

  ret_arr = np.diff(array,k)
  ret_arr = np.true_divide(ret_arr,(h_x ** k))
  return ret_arr
```

```
def get_x_array_derive_k_dimension(original_array,k):
  if (k==0):
    return original_array
    
  return original_array[:len(original_array) - k]
```

```
#draw k-th derive

k=2

X=get_x_array_derive_k_dimension(x,k)
Y=get_y_array_derive_k_dimension(y,k)

my_fig_size = 10
fig = plt.figure(figsize=(my_fig_size+5,my_fig_size))

ax=fig.add_subplot(111)

ax.set_title('  ')
ax.plot(X, Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

```
def search_array_index_of_variable_t(array,value):
  for i in range(len(array)):
    if (array[i]<=value):
      continue
    else:
      return i-1
  return len(array)-1
```

```
#calculate fi_k(t)
def calculate_fi_k(k,t):
  X=get_x_array_derive_k_dimension(x,k)
  Y=get_y_array_derive_k_dimension(y,k)

  #define index of X (t) array
  index = search_array_index_of_variable_t(X,t)

  return Y[index]
```

```
#x^2k / (2k)!
def calculate_intern_sum_k_th_multiplyer(x,k):
  return Decimal(( x**(2*k) )) / Decimal(( math.factorial(2*k) ))
```

```
#calculate u(x,t) in dot x_i
def calculate_U_in_dot(x_array,index,num_of_iterations,t_const):
  sum_in_dot_x_i =0
  for k_iteration in range(num_of_iterations + 1):
    sum_in_dot_x_i += calculate_fi_k(k_iteration,t_const) * float( calculate_intern_sum_k_th_multiplyer(x_array[index],k_iteration) ) 
  
  return sum_in_dot_x_i
```
