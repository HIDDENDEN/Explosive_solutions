```
# importing numpy 
import numpy as np 
np.set_printoptions(linewidth=np.inf)# no extra newlines in output
import math
import matplotlib.pyplot as plt
```

```
#calculate fi(t)

def fi_t(point):
  return math.exp( -1/(point*point) )
```

```
#data to be plotted
segment_len = 4
start_x=0.01
h_x=0.01
array_len = int(segment_len/h_x)


x = np.arange(start_x, start_x + (segment_len), h_x)
# print(x)
print(len(x))
y=np.zeros(array_len)

for i in range(array_len):
  y[i]=fi_t(x[i])
print(y)
print(len(y))
```

```
my_fig_size = 10
fig = plt.figure(figsize=(my_fig_size+5,my_fig_size))

plt.subplot(2, 2, 1)
plt.plot(x, y)

# plt.subplot(2, 2, 2)
# plt.plot(x, y)

# plt.subplot(2, 2, 3)
# plt.plot(x, y)

# plt.subplot(2, 2, 4)
# plt.plot(x, y)

plt.show()
```

```
y_d1 = np.diff(y,n=1)
y_d1 = np.true_divide(y_d1,h_x)

my_fig_size = 10
fig = plt.figure(figsize=(my_fig_size+5,my_fig_size))

ax=fig.add_subplot(111)

ax.set_title('  ')
ax.plot(x[:len(x)-1], y_d1)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# remove the ticks from the top and right edges
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plt.subplot(2, 2, 1)
# plt.plot(x[:len(x)-1], y_d1)

# # plt.subplot(2, 2, 2)
# # plt.plot(x, y)

# # plt.subplot(2, 2, 3)
# # plt.plot(x, y)

# # plt.subplot(2, 2, 4)
# # plt.plot(x, y)

# plt.show()

print(x)
print(x[:(len(x)-1)])
```

```
def get_y_array_derive_k_dimension(array,k):
  ret_arr = np.diff(array,k)
  ret_arr = np.true_divide(ret_arr,h_x ** k)
  return ret_arr
```

```
def get_x_array_derive_k_dimension(original_array,k):
  return original_array[:len(original_array) - k]
```

```
#draw k-th derive

k=2

X=get_x_array_derive_k_dimension(x,k)
Y=get_y_array_derive_k_dimension(y,k)

my_fig_size = 10
fig = plt.figure(figsize=(my_fig_size+5,my_fig_size))

ax=fig.add_subplot(111)

ax.set_title('  ')
ax.plot(X, Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

```
def draw_k_th_derive(k):

  X=get_x_array_derive_k_dimension(x,k)
  Y=get_y_array_derive_k_dimension(y,k)

  my_fig_size = 10
  fig = plt.figure(figsize=(my_fig_size+5,my_fig_size))

  ax=fig.add_subplot(111)

  ax.set_title('  ')
  ax.plot(X, Y)
  ax.spines['left'].set_position('zero')
  ax.spines['right'].set_color('none')
  ax.spines['bottom'].set_position('zero')
  ax.spines['top'].set_color('none')
  ax.xaxis.set_ticks_position('bottom')
  ax.set_xlim([0,2.0])
  ax.yaxis.set_ticks_position('left')
```

```
draw_k_th_derive(16)
```
